## Python Usage Analysis

We analyze how many developers use Python and calculate its popularity among respondents.


```python
worked_with_python = df['LanguageHaveWorkedWith'].str.contains('Python',case=False,na=False).sum()
total = df.shape[0]
percentage=(worked_with_python / total) * 100
print (f'Відсоток респондентів: {round(percentage,1)}%')
```

## Learning Methods Analysis

```python
learned_with_online_courses = df['LearnCode'].str.contains('online courses',case=False,na=False).sum()
learned_with_online_courses
```

## Salary Analysis by Country (Python Developers) 

We analyze salaries of developers who use Python across different countries.


```python
df.worked_with_python = df[df['LanguageHaveWorkedWith'].str.contains('Python',case=False,na=False)]
df.worked_with_python = df.worked_with_python.dropna(subset=['ConvertedCompYearly'])
py_progers_by_country = df.worked_with_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).round()
py_progers_by_country
```

## Education of Top Paid Developers

This section explores the education level of the highest-paid developers.


```python
df_sorted = df.sort_values('ConvertedCompYearly', ascending =False)
df_clean = df.dropna(subset=['ConvertedCompYearly'])
top5 = df_clean.sort_values('ConvertedCompYearly', ascending=False).head(5)
top5[['EdLevel', 'ConvertedCompYearly']]
```

## Python Usage by Age Group

Here we analyze how Python popularity varies across different age groups.


```python
category_age = set(df['Age'].unique())
category_age
```

```python
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python',case=False,na=False)

grouped = df.groupby('Age')
python_count = grouped['worked_with_python'].sum()
total = grouped['worked_with_python'].count()
percentage = (python_count / total * 100).round(1)
percentage


#result = (df.groupby('Age')['worked_with_python'].mean()* 100).round(1)
#result
```

## Top Industries Among High-Paid Remote Workers

This section identifies the most common industries among highly paid remote developers.


```python
Industry_uniq = df['Industry'].unique().shape[0]
Industry_uniq
```

```python
df['Industry'].unique()
```

```python
# рахую порог 0.75 і враховуючи дві умови (тих хто заробляють більше, ніж 75% + працюють віддалено)вже сортуєм респондентів 
q75 = df['ConvertedCompYearly'].quantile(0.75)

result = (
    df[
        (df['ConvertedCompYearly'] > q75) &
        (df['RemoteWork'] == 'Remote')
    ]['Industry']
    .value_counts(dropna=False)
)

result
```

## Python Usage by Age Group

This chart shows how Python usage varies across different age groups.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

percentage.plot(kind='bar')

plt.title('Python Usage by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage')

plt.xticks(rotation=30, ha='right')
plt.tight_layout()

plt.show()
```

## Top Countries by Average Python Developer Salary

This chart presents the top countries with the highest average salaries for Python developers.  
It highlights significant differences in compensation across regions.


```python
top_countries = py_progers_by_country.sort_values('mean', ascending=False).head(10)

plt.figure(figsize=(8,5))

top_countries['mean'].plot(kind='bar')

plt.title('Top 10 Countries by Average Python Developer Salary')
plt.ylabel('Average Salary')

plt.xticks(rotation=30, ha='right')
plt.tight_layout()

plt.show()
```

**Insight:** Salary levels vary significantly between countries.
